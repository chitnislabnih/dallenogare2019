```python
#load libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
```

```python
#load the dataframe
df = pd.read_csv("actin_flow.csv")
```

```python
#statistics

print(df.mean())
print(df.std())
```

```python
#make the graph

fig, ax = plt.subplots(1,1,figsize=(8,2), dpi= 800)


myPlot = sns.swarmplot(x=df["Velocity"], palette=["black", "black"], marker = "o", alpha = 0.6, size=8)


            
#make the boxplot. The "showfliers=False removes the outliers from the boxplot, since they're going to be on the swarmplot anyway
myPlot = sns.boxplot(x=df["Velocity"], color="white", showfliers=False)

for i,box in enumerate(ax.artists):
    box.set_edgecolor('black')
    box.set_facecolor('white')

plt.setp(ax.lines, color="0")


# palette=["white", "black"],

#overlay a swarmplot


#ax.set_ylabel('text', fontdict={'fontsize' : 16})

#set the axis names
#plt.xlabel("condition")
#plt.ylabel("retrograde actin flow velocity $(\mu m.min^{-1})$")

ax.set_xlabel('retrograde actin flow velocity $(\mu m.min^{-1})$', fontdict={'fontsize' : 16})



plt.ylim(-1,1)

plt.xlim(0,12)
plt.show()



#save the plot
figure = myPlot.get_figure()
figure.savefig("retrograde_actin_flow.png",dpi=800, bbox_inches='tight')
```
