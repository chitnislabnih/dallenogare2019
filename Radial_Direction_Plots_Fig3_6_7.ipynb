<h1> Radial Direction Of Protrusions </h1>

<h2> 1. Imports and Data Loading </h2>

```python
# Imports and Settings
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams.update({'font.size': 30})
```

```python
# Load the data

untreated_angle = pd.read_csv("untreated_angle.csv")

apical_intact_angle = pd.read_csv("apical_intact_angle.csv")
apical_skinned_angle = pd.read_csv("apical_skinned_angle.csv")
apical_after_healing_angle = pd.read_csv("apical_after_healing_angle.csv")

basal_intact_angle = pd.read_csv("basal_intact_angle.csv")
basal_skinned_angle = pd.read_csv("basal_skinned_angle.csv")
basal_after_healing_angle = pd.read_csv("basal_after_healing_angle.csv")

apical_DMSO_angle = pd.read_csv("apical_DMSO_angle.csv")
apical_SU5402_angle = pd.read_csv("apical_SU5402_angle.csv")

basal_DMSO_angle = pd.read_csv("basal_DMSO_angle.csv")
basal_SU5402_angle = pd.read_csv("basal_SU5402_angle.csv")

matrigel_angle = pd.read_csv("matrigel_angle.csv")

basal_matrigel_angle = pd.read_csv("basal_matrigel_angle.csv")
```

```python
# Define the plot function

#takes a dataframe. Column 1 of the dataframe is the condition, column 2 is the angle, from 0 to 360.
#optional params saveName (string, the name for the save file, and folder. If 'none' will just display the plot) and bin_size (int, default = 10 degrees)

def radial_plot(df, saveName = None, bin_size = 10):
    plt.clf()   
    
    degrees = df.iloc[:,1]
    radians = np.deg2rad(degrees)

    a , b=np.histogram(degrees, bins=np.arange(0, 360+bin_size, bin_size))
    centers = np.deg2rad(np.ediff1d(b)//2 + b[:-1])

    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='polar')
    ax.bar(centers, a, width=np.deg2rad(bin_size), bottom=0.0, color='.5', edgecolor='k')
    ax.set_theta_zero_location("W")
    ax.set_theta_direction(-1)
    
    
    #change this to set the axis size (default is 0-50)
    ax.set_ylim(0,50)
 
    
    ax.grid(color='gray', linestyle='-', linewidth=2)
    ax.set_axisbelow(True)
    ax.tick_params(axis='x', which='major', pad=15)
    
    
    if saveName != None:
        plt.savefig(saveName, bbox_inches='tight', dpi=800)
    else:
        plt.show()
    
```


<h2> 2. Untreated Embryos </h2>

```python
# Untreated Embryos
radial_plot(untreated_angle)
```
