```python
#load libraries
import seaborn as sns
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import the data
df = pd.read_csv("cell_positions.csv")
```

```python
#split the data

all_nm = df[df['type']=='nm']
all_cell = df[df['type']=='cell']
all_sheath = df[df['type']=='sheath']
```

```python
#plot one embryo as an example

_19_nm = df[df['subtype']=='image_19_nm']
_19_cell = df[df['subtype']=='image_19_cell']
_19_sheath = df[df['subtype']=='image_19_sheath']


f, ax = plt.subplots(figsize=(8, 2))
ax.set_aspect('equal')



ax = sns.kdeplot(_19_cell['x'], _19_cell['y'],
                cmap="Greys", shade=True, shade_lowest=False, bw=1)

ax = sns.kdeplot(_19_nm['x'], _19_nm['y'],
                cmap="Blues", shade=False, shade_lowest=False, bw=2)

ax = sns.kdeplot(_19_sheath['x'], _19_sheath['y'],
                cmap="Reds", shade=False, shade_lowest=False, bw=1)

ax.set_title("19")
ax.set_ylim(25,-25)
ax.set_xlim(-130,80)

figure = ax.get_figure()
#figure.savefig("image19.png",dpi=2000)
```

```python
#average together all embryos and plot as heatmap

f, ax = plt.subplots(figsize=(8, 2))
ax.set_aspect('equal')


ax = sns.kdeplot(all_cell['x'], all_cell['y'],
                cmap="Greys", shade=True, shade_lowest=False, bw=3, alpha=0.8)


ax = sns.kdeplot(all_sheath['x'], all_sheath['y'],
                cmap="Reds", shade=True, shade_lowest=False, bw=3, alpha=0.8)

ax = sns.kdeplot(all_nm['x'], all_nm['y'],
                 shade=False, shade_lowest=False, bw=5, cmap="Blues")


ax.set_title("all")
ax.set_ylim(25,-25)
ax.set_xlim(-130,80)

figure = ax.get_figure()
#figure.savefig("combined_new",dpi=2000)
```
