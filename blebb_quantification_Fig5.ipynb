```python
#load libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
params = {'legend.fontsize': 28,
          'legend.handlelength': 2}
```

```python
#load the data
df = pd.read_csv("blebb_data.csv")
df = pd.DataFrame(df)
df
```

```python
#make the plot

sns.set(style="darkgrid")

%matplotlib inline

sns.set(font_scale=1.5)


# Plot the responses for different events and regions
myPlot = sns.lineplot(x="timepoint", y="membrane",
             data=df,  color="red")

#sns.lineplot(x="timepoint", y="ratio",
#            data=df)

myPlot = sns.lineplot(x="timepoint", y="lifeact",
             data=df, color="green")


plt.legend(labels=['Lyn-Cardinal', 'LifeAct-Citrine'], loc='upper left')
plt.xlim(0, 30)


#set the axis names
plt.xlabel("time (s)")
plt.ylabel("normalized fluorescence")

plt.gcf().subplots_adjust(left=0.15)
plt.gcf().subplots_adjust(bottom=0.15)


figure = myPlot.get_figure()
figure.savefig("bleb_quantification",dpi=1600, bbox='tight')
```
